```python
import pandas as pd
import os
```

```python
# Get a list of all csv files
directory = "exchange"
csv_files = [directory + "/" + file for file in os.listdir(directory) if file.endswith(".csv")]

# Create dataframe of data
results = pd.concat(map(pd.read_csv, csv_files))

stop = list(results.columns).index("validation_mse")


# Get a set of all unique configurations
configurations = list(set(tuple(list(i[1].values)) for i in results.iloc[:,3:stop].iterrows()))

# Store aggregated results in new dataframe
columns = list(results.columns[3:stop])
for col in results.columns[stop:]:
    columns.append(col)
    columns.append(col + "_std")
aggregated_results = pd.DataFrame(columns=columns)

# Iterate over all configurations
for config in configurations:
    aggregated_results.loc[len(aggregated_results)] = {
        **dict(zip(results.columns[3:stop], config)),
        **dict(zip(results.columns[stop:], results[(results.iloc[:,3:stop] == config).all(1)].iloc[:,stop:].mean())),
        **dict(zip([i + "_std" for i in results.columns[stop:]], results[(results.iloc[:,3:stop] == config).all(1)].iloc[:,stop:].std())),
    }

# Sort data based on validation variogram score
aggregated_results = aggregated_results.sort_values("validation_variogram")
```
